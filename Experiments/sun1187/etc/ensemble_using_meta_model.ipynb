```python
import pandas as pd
import numpy as np
```

# Meta ensemble

```python
catboost = pd.read_csv('catBoost.csv')
catboost.drop('Unnamed: 0', axis=1, inplace=True)
catboost
```

```python
lgbm = pd.read_csv('lgbm.csv')
lgbm.drop('Unnamed: 0', axis=1, inplace=True)
lgbm
```

```python
lightgcn = pd.read_csv('/opt/ml/input/code/dkt_sh/dkt/lightgcn/lightgcn_train_data.csv')
lightgcn_pred = pd.read_csv('/opt/ml/input/code/dkt_sh/dkt/lightgcn/output/best_auc_submission.csv')

lightgcn = pd.concat([lightgcn, lightgcn_pred], axis=1)
lightgcn
```

```python
catboost = catboost.sort_values('ori_index')
lightgcn = lightgcn.sort_values('ori_index')

lightgcn.id = lightgcn.ori_index
catboost.id = catboost.ori_index

catboost = catboost.reset_index()
catboost.drop('index', inplace=True, axis=1)
```

```python
models_output = pd.DataFrame()
models_output['catboost'] = catboost['prediction']
models_output['lightgcn'] = lightgcn['prediction']
models_output['lgbm'] = lgbm['prediction']
models_output['answerCode'] = lgbm.answerCode
```

```python
models_output
```

# 평균 합

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
y = models_output.pop('answerCode')
X_train, X_test, y_train, y_test = train_test_split(
    models_output, y, test_size=0.3, random_state=42)
```

```python
train_mean_hap = (X_train.catboost + X_train.lightgcn + X_train.lgbm)/3
ensem_train_acc = accuracy_score(y_train, np.where(train_mean_hap >= 0.5, 1, 0))
ensem_train_auc = roc_auc_score(y_train, train_mean_hap)

test_mean_hap = (X_test.catboost + X_test.lightgcn + X_test.lgbm)/3
ensem_valid_acc = accuracy_score(y_test, np.where(test_mean_hap >= 0.5, 1, 0))
ensem_valid_auc = roc_auc_score(y_test, test_mean_hap)
```

```python
print(f'TRAIN MEAN_HAP AUC : {ensem_train_auc} ACC : {ensem_train_acc}')
print(f'VALID MEAN_HAP AUC : {ensem_valid_auc} ACC : {ensem_valid_acc}')
```

```python
ensem_train_acc = accuracy_score(y_train, np.where(X_train.catboost >= 0.5, 1, 0))
ensem_train_auc = roc_auc_score(y_train, X_train.catboost)

ensem_valid_acc = accuracy_score(y_test, np.where(X_test.catboost >= 0.5, 1, 0))
ensem_valid_auc = roc_auc_score(y_test, X_test.catboost)

print(f'TRAIN CATBOOST AUC : {ensem_train_auc} ACC : {ensem_train_acc}')
print(f'VALID CATBOOST AUC : {ensem_valid_auc} ACC : {ensem_valid_acc}')
```

# 메타 모델

```python
# 메타러닝
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

meta_model = LogisticRegression()
#meta_model = GaussianNB()
#meta_model = DecisionTreeClassifier()
meta_model.fit(X_train, y_train)

meta_train_pred = meta_model.predict_proba(X_train)

meta_train_pred = meta_train_pred[:,-1]
meta_train_acc = accuracy_score(y_train, np.where(meta_train_pred >= 0.5, 1, 0))
meta_train_auc = roc_auc_score(y_train, meta_train_pred)

## valid
meta_valid_pred = meta_model.predict_proba(X_test)
meta_valid_pred = meta_valid_pred[:,-1]
meta_valid_acc = accuracy_score(y_test, np.where(meta_valid_pred >= 0.5, 1, 0))
meta_valid_auc = roc_auc_score(y_test, meta_valid_pred)
```

```python
print(f'TRAIN MEAN_HAP AUC : {ensem_train_auc} ACC : {ensem_train_acc}')
print(f'VALID MEAN_HAP AUC : {ensem_valid_auc} ACC : {ensem_valid_acc}\n')

print(f'TRAIN MEAN_HAP AUC : {meta_train_auc} ACC : {meta_train_acc}')
print(f'VALID MEAN_HAP AUC : {meta_valid_auc} ACC : {meta_valid_acc}')
```
